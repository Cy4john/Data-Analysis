```python
# Function to calculate the area of a rectangle
def area_rectangle(length, width):
    area = length * width
    return area

# Function to calculate the factorial of a number
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)

# Function to check if a number is prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Function to calculate the sum of all elements in a list
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

# Testing the functions
print(area_rectangle(5, 10))
print(factorial(5))
print(is_prime(7))
print(sum_list([1, 2, 3, 4, 5]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.arange(0.0, 2.0, 0.01)
y = np.sin(2 * np.pi * x)

# Line plot
plt.plot(x, y)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()

# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Scatter Plot')
plt.show()

# Bar chart
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
values = [20, 35, 30, 25]

plt.bar(labels, values)
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Sale')
plt.show()

# Pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 30, 15, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Programming Language Popularity')
plt.show()
```

```python
#This program uses the Matplotlib library to create four different types of plots for a biscuits company:

#A line plot of monthly sales performance for the year.
#A bar chart of sales by product.
#A pie chart of sales by region.
#A scatter plot of advertising spend vs. sales.

import matplotlib.pyplot as plt
import numpy as np

# Sales data for the year
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12000, 15000, 20000, 18000, 25000, 22000, 28000, 30000, 27000, 32000, 35000, 40000]

# Line plot of monthly sales
plt.plot(months, sales)
plt.xlabel('Month')
plt.ylabel('Sales (in USD)')
plt.title('Monthly Sales Performance')
plt.grid(True)
plt.show()

# Bar chart of sales by product
products = ['Cookies', 'Crackers', 'Wafers', 'Biscuits']
sales_by_product = [50000, 35000, 40000, 45000]

plt.bar(products, sales_by_product)
plt.xlabel('Product')
plt.ylabel('Sales (in USD)')
plt.title('Sales by Product')
plt.show()

# Pie chart of sales by region
regions = ['North', 'South', 'East', 'West']
sales_by_region = [40000, 35000, 45000, 30000]

plt.pie(sales_by_region, labels=regions, autopct='%1.1f%%')
plt.title('Sales by Region')
plt.show()

# Scatter plot of advertising spend vs. sales
advertising_spend = [10000, 12000, 14000, 16000, 18000, 20000, 22000, 24000, 26000, 28000]
sales = [12000, 15000, 20000, 18000, 25000, 22000, 28000, 30000, 27000, 32000]

plt.scatter(advertising_spend, sales, alpha=0.5)
plt.xlabel('Advertising Spend (in USD)')
plt.ylabel('Sales (in USD)')
plt.title('Advertising Spend vs. Sales')
plt.show()
```
